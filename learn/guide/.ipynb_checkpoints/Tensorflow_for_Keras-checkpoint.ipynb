## DNN_for_imageClassification 

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

### 载入数据 

```python

fashion_data = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_data.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

### 预清洗数据

```python

plt.imshow(train_images[0])
plt.colorbar()

train_images = train_images/255.0
test_images = test_images/255.0

plt.figure(figsize=(10,10))

for i in range (25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i]])
```

### 构建模型 

#### 设置层 

```python
# 28*28 二维数组转化成一维784 数组
# 两个网络层训练，784 —> 128 ->10
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(128,activation=tf.nn.relu),
    keras.layers.Dense(10,activation=tf.nn.softmax)
])
```

#### 编译模型 

```python
# 增加模型训练的参数
model.compile(optimizer = tf.train.AdamOptimizer(),
             loss = "sparse_categorical_crossentropy",
             metrics = ['accuracy'])
```

### 训练模型 

```python
# 载入训练集，每一批数据训练结束后，loss 都会提升，大概是在88%
model.fit(train_images,train_labels,epochs=6)
```

### 评估准确率 

```python
# 测试集来评估准确率率
test_loss, test_acc = model.evaluate(test_images, test_labels)
# 过拟合 ：测试集效果比训练差3个百分点
print('Test accuracy:', test_acc)
```

### 预测 

```python
predictions = model.predict(test_images)

# DNN 输出，10个类别各自的概率
predictions[0]
# 使用概率最大的类别当作最后的预测标签；softmax
np.argmax(predictions[0])
# test 真实的label
test_labels[0]
```

```python
def plt_img(i,predictions_array,true_labels,imgs):
    prediction_array,true_label,img = predictions_array[i],true_labels[i],imgs[i];
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img,cmap=plt.cm.binary)
    
    prediction_label = np.argmax(prediction_array)
    
    # 判断预测是否正确
    if prediction_label == true_label:
        color = 'blue'
    else:
        color = 'red'
        
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[prediction_label],
                                100*np.max(prediction_array),
                                class_names[true_label]),
                                color=color)
    
def plt_value_array(i,predictions_array,true_labels):
    prediction_array,true_label = predictions_array[i],true_labels[i]
    plt.xticks([])
    plt.yticks([])
    thisplot = plt.bar(range(10),prediction_array,color="#777777")
    plt.ylim(0,1)
    prediction_label = np.argmax(prediction_array)
    
    thisplot[prediction_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
i=555
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt_img(i,predictions,test_labels,test_images)

plt.subplot(1,2,2)
plt_value_array(i,predictions,test_labels)
```

```python
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plt_img(i, predictions, test_labels, test_images)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plt_value_array(i, predictions, test_labels)
```
